#Seaborn


*  Seaborn is a Python data visualization library built on top of Matplotlib.




*   It provides beautiful, easy-to-use, and informative statistical graphics.


*   Seaborn works well with Pandas DataFrames and supports advanced plots like heatmaps, boxplots, and violin plots with simple commands.



```python
import seaborn as sns
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#line Plot
A Line Plot is used to display data points connected by straight lines. It shows trends over time or sequences.

```python
sns.lineplot(x="Rollno", y="Marks", data=sample_df)
plt.title("Students Marks")
```

#Dist plot
A Dist Plot is used to visualize the distribution of a dataset. It shows both the histogram (frequency) and the KDE (Kernel Density Estimation) to understand how the data is spread.

```python
 sns.distplot(df['time_spend_company'])
```

```python
sns.distplot(df['left'])
```

```python
sns.distplot(df['average_montly_hours'])
```

```python
df.describe()
```

```python
bins = [2,3,4,5,6,7,8,9,10]
sns.distplot(df['time_spend_company'],bins=bins)
plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'],bins=bins,kde=False)
```

```python
sns.distplot(df['time_spend_company'],bins=bins,hist=False)
```

```python
#orient
sns.barplot (y = 'class' ,x = 'fare', data = titanic_df,hue = 'sex',
             palette = 'inferno',orient = 'h')                          #for horizontal orientation x should be a numerical value
```

```python
#estimator
sns.barplot(x = 'class', y = 'fare', data = titanic_df,hue = 'sex',
            palette = 'inferno',estimator = np.max)
```

```python
#confidence interval
sns.barplot(x = 'class', y = 'fare', data = titanic_df,
            hue = 'sex',palette = 'inferno',
            ci=100,                                    #ci ranges from 0 to 100
            errcolor = 'red',errwidth = 10)
```

```python
#confidence interval
sns.barplot(x = 'class', y = 'fare', data = titanic_df,
            hue = 'sex',palette = 'inferno',
            ci=100,
            errcolor = '#7777da',errwidth = 10)
```

```python
#saturation
sns.barplot(x = 'class', y = 'fare', data = titanic_df,
            hue='sex',palette = 'inferno',saturation = 0.1)       #more saturatio more colorfull
```

```python
#saturation
sns.barplot(x = 'class', y = 'fare', data = titanic_df,
            hue='sex',palette = 'inferno',saturation = 0.9)       #more saturatio more colorfull
```

#Heatmaps

```python
flights_df=sns.load_dataset("flights")
flights_df.head()
```

```python
flights_df=flights_df.pivot(index='month', columns='year', values='passengers')
flights_df.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(flights_df)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(flights_df,annot = True,fmt = 'd')
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(flights_df,annot = True,fmt = 'd',
            linecolor='k',linewidths=5)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(flights_df,annot = True,fmt = 'd',
            linecolor='k',linewidths=5, cmap='Blues')
```
